1.What is a random variable in probability theory?

2.What are the types of random variables?

3.What is the difference between discrete and continuous distributions?
  - **Discrete** **Distribution**: Describes probability of a discrete random variable.

4.What are probability distribution function (PDF)?
  - For a continuous random variable, a PDF specifies the probability density at each point. The total area under the curve equals 1.

5.How do comulative distribution functions (CDF) Differe form probability distribution function (PDF)?

6.What is a discrete uniform distribution?

7.What are the key properties of a Bernoulli distribution?
  - Only two outcome: success (1) or failure (0).
  - Single trial with probability p for success and 1 - p for failure.
  - mean = p , variance = P(1 - p)

8.What is the Binomial distribution, and how is it used in probability?
  - The binomial distribution describes the number of successes un n independent Bernouli trial with probability p of success.

9.What is the possion distribution and where is it applied?

10.What is a continuous uniform distribution?

11.What are the characteristics of a normal distribution?
  - Symmetrical, bell-shaped curve.
  - Defined by mean (μ) and standerd deviation ( σ).
  - Most values cluster around the mean.

12.What is the standard normal distribution, and why is it important?
  - A normal distribution with a mean of 0 and a standard deviation of 1. It simplifies probability calculations and standardizes data.

13.What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - The CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population's distribution.

14.How does the Central Limit Theorem relate to the normal distribution?
  - The CLT ensures that averages of large samples follow a normal distribution, enabling statistical inference using normal probability.

15.What is the application of Z statistics in hypothesis testing?
  - Z statistics determine how far a sample mean deviates from the population mean under the full hypothesis, measured in standard deviations.

16.How do you calculate a Z-score, and what does it represent?
  -  Z = (X - μ) / σ

  where:

 - Z is the Z-score
 -  X is the individual data point
 -  μ is the population mean
 -   σ is the population standard deviation



17.What are point estimates and interval estimates in statistics?
  - **Point Estimate**: A single value (e.g., sample mean) that estimates a population parameter.
  - **Interval Estimate:** A range of values (confidence interval) that likely contains the parameter.

18.What is the significance of confidence interval in statistical analysis?
- Confidence intervals provide a range of plausible values for a population parameter, reflecting estimation uncertainty.

19.What is the relationship between a Z-score and a confidence interval?

20.How are Z-scores used to compare different distribution?

21.What are the assumptions for applying the Central Limit Theorem?

22.What is the concept of expected value in a probability distribution?

2.Generate aa discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import matplotlib.pyplot as plt
import numpy as np


low, high = 1, 6

x = np.arange(low, high + 1)
pmf = [1 / (high - low + 1)] * len(x)

plt.stem(x, pmf, basefmt=" ")
plt.title('Discrete Uniform Distribution PMF')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```

3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(p, x):
    if x in [0, 1]:
      return p if x == 1 else 1 - p
    else:
      return 0

p = 0.6
x = 1
pdf_value = bernoulli_pdf(p, x)
print(f"PDF at x = {x}: {pdf_value}")
```

9.Write a Python function to calculate Z-scores from a dataset and plot them.

```python
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

data = np.random.normal(10, 2, size=1000)
z_scores = calculate_z_scores(data)
plt.hist(z_scores, bins=30, edgecolor="black", alpha=0.7)
plt.title('Histogram of Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.show()
```

16.Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).

```python
from scipy.stats import norm

def plot_standard_normal_distribution(mean=0, std=1):
    x = np.linspace(-4, 4, 100)
    pdf = norm.pdf(x, loc=mean, scale=std)

    plt.plot(x, pdf, label='Standard Normal Distribution')
    plt.title('Standard Normal Distribution')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()

plot_standard_normal_distribution()
```

18.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
def calculate_z_score(data, point):
    mean = np.mean(data)
    std = np.std(data)
    z_score = (point - mean) / std
    return z_score

data = np.random.normal(10, 2, 100)
point = 15
z_score = calculate_z_score(data, point)
print("Z-score:", z_score)

from scipy.stats import norm

p_value = norm.cdf(z_score)
print("P-value (comparison to standard normal):", p_value)
```

19.Implement hypothesis testing using Z-statistics for a sample dataset.

```python
def z_test(sample_mean, population_mean, sample_std, n, alpha):
    z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
    return z_score

sample_mean = 75
population_mean = 70
sample_std = 5
n = 100
alpha = 0.05

z_stat = z_test(sample_mean, population_mean, sample_std, n, alpha)

p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print("Z-statistics:", z_stat)
print("P-value:", p_value)
```

20.Create a confidence interval for a dataset using Python and interpret the result.

```python
import scipy.stats as stats

data = np.random.normal(10, 2, 100)
mean = np.mean(data)
std = np.std(data, ddof=1)
confidence_level = 0.95

h = stats.t.ppf((1 + confidence_level) / 2, df=len(data) - 1) * (std / np.sqrt(len(data)))
confidence_interval = (mean - h, mean + h)

print("05% Confidence Interval:", confidence_interval)
```
